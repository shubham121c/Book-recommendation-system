popularity based recommender system
-will show top 50 books having highest avg ratings
-should have min 250 ratings

```python
# joining 2 tabales
# to get ratings and book info
rating_book=ratings.merge(books, on='ISBN')
rating_book.head()
```

```python
nrating = rating_book.groupby('Book-Title').count()['Book-Rating'].reset_index()
nrating.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
nrating.head()
```

```python
arating = rating_book.groupby('Book-Title').mean()['Book-Rating'].reset_index()
arating.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
arating.head()
```

```python
popular= nrating.merge(arating, on='Book-Title')
popular.head()
```

```python
popular=popular[popular['num_ratings']>=250].sort_values('avg_ratings', ascending=False)
popular.shape
```

```python
popular= popular.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
popular
```

collobrative filtering based recommender system

```python
x=rating_book.groupby('User-ID').count()['Book-Rating']>200
iusers=x[x].index
# filtering out indexes of users who hav atleaset rated 200 books
```

```python
filtered_rating=rating_book[rating_book['User-ID'].isin(iusers)]
# filtering out the rows where these iusers have rated the books
```

```python
abc=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=abc[abc].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings['Image-URL-M'][63]
```

```python
pt=final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
similarity_score.shape
```

```python
def recommend(book_name):
    #fetching index of input book_name
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
   
```
